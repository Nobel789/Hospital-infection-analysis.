Ward  infection  Pattern detection by Heatmap

Step 1: Sample Healthcare Data

We’ll use your infection example:

```python
import pandas as pd

data = {
    "Ward": ["ICU", "ER", "Pediatrics", "Surgery"],
    "Week 1": [5, 2, 1, 6],
    "Week 2": [8, 3, 2, 9],
    "Week 3": [12, 4, 1, 10]
}

df = pd.DataFrame(data)
df.set_index("Ward", inplace=True)

print(df)
```

Step 2: Create Heatmap (Using Seaborn)

We use seaborn because it makes heatmaps very simple.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure()

sns.heatmap(df, annot=True, cmap="Reds")

plt.title("Hospital Infection Rates Heatmap")
plt.xlabel("Weeks")
plt.ylabel("Hospital Wards")

plt.show()
```

📊 Simple Insight

👉 “Infection rates are increasing over time, with ICU and Surgery showing the highest and fastest-growing infection levels, especially in Week 3.”

🧠 What this means:
ICU: 5 → 8 → 12 🚨 (sharp increase, highest risk)
Surgery: 6 → 9 → 10 ⚠️ (consistently high and rising)
ER: 2 → 3 → 4 🟡 (moderate increase)
Pediatrics: 1 → 2 → 1 🟢 (low and stable)

👉 The main problem areas = ICU and Surgery
